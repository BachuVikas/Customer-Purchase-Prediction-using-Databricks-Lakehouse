# E-Commerce Customer Purchase Prediction using Databricks Lakehouse

## Project Overview

E-commerce companies collect large amounts of transaction data every day.  
However, identifying which customers are likely to purchase again can be difficult without proper data analysis and machine learning models.

This project builds an end-to-end data and AI pipeline using **Databricks Lakehouse** to predict whether a customer is likely to make another purchase.

The solution uses **Medallion Architecture (Bronze → Silver → Gold)** with Delta Lake tables and a machine learning model to generate predictions.

---

## Problem Statement

Marketing teams in e-commerce companies want to target customers who are most likely to purchase again.

Instead of sending promotions to all customers, businesses can increase marketing efficiency by identifying customers with high purchase probability.

The goal of this project is to build a machine learning system that predicts whether a customer will purchase again based on their historical transaction data.

---

## Objectives

The main objectives of this project are:

- Build a scalable data pipeline using Databricks Lakehouse
- Implement Medallion Architecture (Bronze, Silver, Gold layers)
- Perform data cleaning and feature engineering using PySpark
- Train a machine learning model to predict customer purchase behavior
- Track experiments using MLflow
- Generate predictions and store them in Delta tables

---

## Business Impact

The predictions generated by this system can help businesses:

- Improve marketing targeting
- Increase customer retention
- Identify high-value customers
- Optimize marketing campaign costs

## Architecture Overview

The data pipeline follows the Medallion Architecture approach.

Bronze Layer  
Raw transactional data ingested into Delta tables.

Silver Layer  
Cleaned and structured transactional data.

Gold Layer  
Customer-level aggregated features used for machine learning.

Machine Learning Layer  
Train a classification model to predict purchase likelihood.

Prediction Layer  
Store predictions in a Gold table for analytics and business use.
